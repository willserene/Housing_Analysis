```python
mls = []
sale_price = []
close_date =[]
address = []
zip_code = []
total_sq = []
lot_size = []
linebreak = "-----------------------------------------------------------------------------------------"
count = 1
for i in range(len(sales_json)):
    print(linebreak)
    print("Beginning Data Retreival Process")
    print(linebreak)

    try:
        mls.append(sales_json[i][0])
        sale_price.append(sales_json[i][1])
        close_date.append(sales_json[i][2])
        address.append(sales_json[i][3])
        zip_code.append(sales_json[i][4])
        total_sq.append(sales_json[i][5])
        lot_size.append(sales_json[i][6])
        print(f"Retreiving Data for House #{count}")
        count += 1
    except:
        print("Bzzzt. Kssschhh. Error. ERROR!")
        pass

print(linebreak)
print("Data Retrieval Complete")
print(linebreak)

print(mls[-1])
print(sale_price[-1])
print(close_date[-1])
print(address[-1])
print(zip_code[-1])
print(total_sq[-1])
print(lot_size[-1])

```

```python
dict = {"MLS ID": mls, "Sale Price": sale_price, "Close Date": close_date, "Address": address, "Zip Code": zip_code,
              "Total Sq. Ft.": total_sq, "Lot Size": lot_size}
```

```python
sales_df = pd.DataFrame(dict)
sales_df
```
