```python
address_url_list = [x.replace('#', '%23') for x in sales_df["Address"]]

address_url_list = [x.replace('VW', 'View') for x in address_url_list]

address_url_list = [x.replace('BL', 'Blvd') for x in address_url_list]

address_url_list = [x.replace('LN', 'Lane') for x in address_url_list]

address_url_list = [x.replace('CR', 'Cir') for x in address_url_list]

address_url_list = [x.replace('HT', 'Heights') for x in address_url_list]

address_url_list = [x.replace('TR', 'Trail;') for x in address_url_list]

address_url_list = [x.replace('DR', 'Drive') for x in address_url_list]

address_url_list = [x.replace('TE', 'Terrace') for x in address_url_list]

address_url_list = [x.replace('WY', 'Way') for x in address_url_list]

address_url_list = [x.replace('PT', 'Point') for x in address_url_list]

address_url_list = [x.replace('GR', 'Grove') for x in address_url_list]

address_url_list = [x.replace('AV', 'Ave') for x in address_url_list]

address_url_list = [x.replace('PW', 'Parkway') for x in address_url_list]

address_url_list = [x.replace('LP', 'Loop') for x in address_url_list]

address_url_list = [x.replace('CT', 'Court') for x in address_url_list]

address_url_list = [x.replace(' ', '%20') for x in address_url_list]

#address_url_list = [x.replace('ST', 'Cir') for x in address_url_list]

print(address_url_list)
```

```python
# Testing out query and finding out json posisitons to look for in for loop
query_url = f"https://maps.googleapis.com/maps/api/geocode/json?address={address_url_list[-1]}&key={g_key}"
print(query_url)    
address_response = requests.get(query_url)
address_json = address_response.json()
print(json.dumps(address_json, indent=4))
```

```python
address_json["results"][0]['geometry']['location']
```

```python
lon = []
lat = []
count = 1
errorlist = []
errorcount = 1
linebreak = "-----------------------------------------------------------------------------------------"

print("Begin Lon Lat Retrieval")
print(linebreak)

for address_url  in address_url_list:
    
    query_url = f"https://maps.googleapis.com/maps/api/geocode/json?address={address_url}%20Colorado%20Springs%20CO&key={g_key}"
    
    try:
        address_response = requests.get(query_url)
        address_json = address_response.json()
        lon.append(address_json["results"][0]['geometry']['location']['lng'])
        lat.append(address_json["results"][0]['geometry']['location']['lat'])
        print(f"Processing {count} of {len(address_url_list)}")
        count += 1
    
    except:
        errorcount += 1
        print(f'Error #{errorcount}')
        errorlist.append(count)
        lon.append('NaN')
        lat.append('Nan')
        count += 1
        pass

print(linebreak)
print("Finished with Lon Lat Retrieval")
print(f"{len(lon)} longitudes and {len(lat)} latitudes were found")
print(f"{errorcount} Errors were found")
print(errorlist)
```
