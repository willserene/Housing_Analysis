```python
# high school names and ranks
high_school = []
high_school_rank = []

count = 1

linebreak = "---------------------------------------------------------------------------------------------"
print("Beginning Data Retrieval")
print(linebreak)

for index,row in sales_df.iterrows():
    
    lat = row["Latitude"]
    
    lng = row["Longitude"]

    url = f"https://api.schooldigger.com/v1.2/schools?st=CO&level=High&nearLatitude={lat}&nearLongitude={lng}&distanceMiles=5&sortBy=distance&appID=52621e3a&appKey=ea14676381f70e7dccb7e08aa287528c"

    try:
        response = requests.get(url)
        school_json = response.json()
        high_school.append(school_json['schoolList'][0]['schoolName'])
        high_school_rank.append(school_json['schoolList'][0]['rankHistory'][0]['rankStars'])
        print(f"Retrieval {count} out of {len(sales_df.index)}: {school_json['schoolList'][0]['schoolName']}")
        count += 1
        
    except:
        print("Error")
        high_school.append("NaN")
        high_school_rank.append("NaN")
        count += 1
        pass
    
print(linebreak)
print("THE END")
    
```

```python
print(len(high_school))
print(len(high_school_rank))
```

```python
highschool_df = pd.DataFrame({"High School": high_school})
highschool_df
```

```python
highschoolrank_df = pd.DataFrame({"High School Rank": high_school_rank})
highschoolrank_df.to_csv("high_school_rank.csv", sep=",", encoding='utf-8', index=False, header=True)
```
